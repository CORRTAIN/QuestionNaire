```python
import pandas as pd # 导入另一个包“pandas” 命名为 pd，理解成pandas是在 numpy 基础上的升级包
import numpy as np #导入一个数据分析用的包“numpy” 命名为 np
import matplotlib.pyplot as plt # 导入 matplotlib 命名为 plt，类似 matlab，集成了许多可视化命令
from scipy import stats #从scipy中导入stats
#jupyter 的魔术关键字（magic keywords）
#在文档中显示 matplotlib 包生成的图形
# 设置图形的风格
%matplotlib inline 
%config InlineBackend.figure_format = 'retina'
```

```python
data = pd.read_excel('~/JupyterThemes/QuestionNaire/优等生调查统计.xlsx')
data #查看数据
```

```python
#赋值
for i in data.columns:
    for j in data.index:
        if data[i][j] == 'A':
            data[i][j] = 1
        if data[i][j] == 'B':
            data[i][j] = 2
        if data[i][j] == 'C':
            data[i][j] = 3
        if data[i][j] == 'D':
            data[i][j] = 4
```

```python
#mean均值，是正态分布的中心，把 数据集中的均值 定义为 mean
mean = data.mean()
mean
```

```python
#S.D.标准差，把数据集中的标准差 定义为 std
std = data.std()
std
```

```python
# 设定 x 轴前两个数字是 X 轴的开始和结束，第三个数字表示步长，或者区间的间隔长度
x = np.arange(1,4,0.1) 
#设定 y 轴，求正态分布函数曲线
y = stats.norm.pdf(x,mean[9],std[9])
plt.plot(x,y)
#画出曲线图
plt.title('Top student Question9')
plt.xlabel('Answer')
plt.ylabel('Probability')
#输出
plt.show()
```
